```python
df["Price Date"] = pd.to_datetime(df["Price Date"])


df = df.sort_values(["Market Name", "Commodity", "Price Date"])
```

```python
df["day"] = df["Price Date"].dt.day
df["month"] = df["Price Date"].dt.month
df["year"] = df["Price Date"].dt.year
df["dayofweek"] = df["Price Date"].dt.dayofweek
```

```python
df["lag1"] = df.groupby(["Market Name","Commodity"])["Modal_Price"].shift(1)
df["lag2"] = df.groupby(["Market Name","Commodity"])["Modal_Price"].shift(2)
df["lag3"] = df.groupby(["Market Name","Commodity"])["Modal_Price"].shift(3)
```

```python
df = df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

le_market = LabelEncoder()
le_commodity = LabelEncoder()

df["Market Name"] = le_market.fit_transform(df["Market Name"])
df["Commodity"] = le_commodity.fit_transform(df["Commodity"])
```

```python
features = ["Market Name","Commodity","Min_Price","Max_Price",
            "day","month","year","dayofweek",
            "lag1","lag2","lag3"]

X = df[features]
y = df["Modal_Price"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
import pandas as pd
from datetime import timedelta

def predict_next_5_days(market, commodity, today_date):

    future_preds = []

    # encode inputs
    market_enc = le_market.transform([market])[0]
    commodity_enc = le_commodity.transform([commodity])[0]

    # get last known records for lag values
    temp = last_data[(last_data["Market Name"] == market_enc) & 
                     (last_data["Commodity"] == commodity_enc)].sort_values("Price Date")

    lag1 = temp.iloc[-1]["Modal_Price"]
    lag2 = temp.iloc[-2]["Modal_Price"]
    lag3 = temp.iloc[-3]["Modal_Price"]

    current_date = pd.to_datetime(today_date)

    for i in range(5):

        future_date = current_date + timedelta(days=i+1)

        row = pd.DataFrame([{
            "Market Name": market_enc,
            "Commodity": commodity_enc,
            "Min_Price": lag1,   # approximation
            "Max_Price": lag1,
            "day": future_date.day,
            "month": future_date.month,
            "year": future_date.year,
            "dayofweek": future_date.dayofweek,
            "lag1": lag1,
            "lag2": lag2,
            "lag3": lag3
        }])

        pred = model.predict(row)[0]

        future_preds.append((future_date.date(), round(pred,2)))

        # shift lags
        lag3 = lag2
        lag2 = lag1
        lag1 = pred

    return future_preds
```
